###날짜를 위한 함수 data_range()  1학년때 시험문제

```python
dates = pd.date_range('20250101', periods=6)
dates
```

```python
dates[5]
```

```python
dates = pd.date_range('20250101', end='20250110')
dates
```

```python
dates = pd.date_range("20250101", periods=6, freq='3H')
dates
```

```python
dates = pd.date_range("20250101", periods=6, freq='W') # W로 하면 날짜(20250101)가 있는 주 의 일요일을 시작으로 나옴
dates
```

```python
dates = pd.date_range("20250101", periods=6, freq='W-SUN')
# W로 하면 날짜(20250101)가 있는 주 의 일요일을 시작으로 나옴
# W-SUN 해도 같음 (다른 요일도 가능)
dates
```

```python
dates = pd.date_range("20250101", periods=6, freq='ME') # ME = 달의 마지막
dates
```

```python
dates = pd.date_range("20250101", periods=6, freq='MS') # MS = 달의 첫날
dates
```

```python
dates = pd.date_range("20250101", periods=6)
dates
```

```python
df = pd.DataFrame(np.random.randn(6,4), index=dates, columns=list('ABCD'))  #dates를 index로 가져옴
df
```
